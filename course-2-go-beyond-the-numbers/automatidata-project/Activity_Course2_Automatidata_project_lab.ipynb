To confirm your conclusion, consider the following experiment:
1. Create a sample of coordinates from a normal distribution&mdash;in this case 1,500 pairs of points from a normal distribution with a mean of 10 and a standard deviation of 5
2. Calculate the distance between each pair of coordinates 
3. Group the coordinates by endpoint and calculate the mean distance between that endpoint and all other points it was paired with
4. Plot the mean distance for each unique endpoint

```python
#BONUS CONTENT

#1. Generate random points on a 2D plane from a normal distribution
#==> ENTER YOUR CODE HERE

# 2. Calculate Euclidean distances between points in first half and second half of array
#==> ENTER YOUR CODE HERE

# 3. Group the coordinates by "drop-off location", compute mean distance
#==> ENTER YOUR CODE HERE

# 4. Plot the mean distance between each endpoint ("drop-off location") and all points it connected to
#==> ENTER YOUR CODE HERE

# 1. Generate random points
test = np.round(np.random.normal(10, 5, (1500, 2)), 1)

# 2. Calculate Euclidean distances
midpoint = len(test)//2
distances = np.sqrt((test[:midpoint, 0]-test[midpoint:, 0])**2 + 
                    (test[:midpoint, 1]-test[midpoint:, 1])**2)

# 3. Group by endpoint, compute mean distance
mean_distances = pd.DataFrame({'endpoint': test[:midpoint, 0], 
                                'distance': distances}).groupby('endpoint')['distance'].mean()

# 4. Plot
sns.barplot(x=mean_distances.index, y=mean_distances, color='steelblue')
plt.title("Mean distance from each endpoint")
plt.xticks(rotation=90)
plt.show()
```

The concept: Regression to the mean and random variation
When you have many groups (like 216 drop-off locations) and calculate a mean for each, some groups will naturally appear high and some low — even if the underlying data is completely random. This is not necessarily meaningful signal, it's just statistical noise.
What the bonus is showing you:
The spiky bar chart of mean trip distances by location looks dramatic — some locations have very high means. But when you generate completely random coordinates and do the same calculation, you get a similarly spiky chart. This raises an important question: is the variation in your real data genuinely meaningful, or is some of it just random noise?
The practical takeaway for your project:
Before concluding that certain drop-off locations have meaningfully longer trips, you need enough data points per location to trust the mean. A location with only 2-3 trips could show a very high mean just by chance. A location with 500 trips showing a high mean is much more reliable.
In plain language:

Don't over-interpret means calculated from small samples. A spiky pattern across many groups can appear even in purely random data.

This is a foundational concept in statistics you'll encounter repeatedly — especially when building models. Keep it in mind.
